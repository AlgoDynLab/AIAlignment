```python
import pandas as pd
```

```python
data = pd.read_csv('/Users/beto/Documents/Projects/AIAlignment/data/agent_conversations.csv')
data.head(2)
```

```python
import os

# Create data directory if it doesn't exist
os.makedirs('data', exist_ok=True)

# Group data by topic_discussion
topic_groups = data.groupby('topic_discussion')

# Create a txt file for each topic
for topic, group in topic_groups:
    filename = f'data/{topic}.txt'
    
    # Create content by concatenating character and comment
    lines = []
    for _, row in group.iterrows():
        line = f"{row['character']}:{row['comment']}"
        lines.append(line)
    
    # Write lines to file
    with open(filename, 'w', encoding='utf-8') as f:
        f.write('\n'.join(lines))
```
